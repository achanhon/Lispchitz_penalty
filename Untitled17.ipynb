```python
import torch

def createSample():
  x = 0.9*torch.rand(3,128,128)
  y = torch.zeros(128,128)

  # add signal
  for k in range(3):
    tmp = torch.rand(2)*120
    j,i = int(tmp[0]),int(tmp[1])
    x[0,j:j+7,i]=1
    x[0,j:j+7,i+7]=1
    x[1,j,i:i+7]=1
    x[1,j+7,i:i+7]=1
    y[j+3][i+3]=1

  return x,y

import matplotlib.pyplot as plt

x,y = createSample()

y = torch.stack([y,y,y],dim=0)
xy = torch.cat([x,y],dim=-1)
xy = xy.cpu().numpy().transpose(1,2,0)

plt.imshow(xy)
plt.axis('off')
plt.show()
```

```python
class LittleUnet(torch.nn.Module):
  def __init__(self):
    super(LittleUnet, self).__init__()
    self.c11 = torch.nn.Conv2d(3, 32,kernel_size=3,padding=1)
    self.c12 = torch.nn.Linear(32, 32,kernel_size=3,padding=1)

    self.c21 = torch.nn.Linear(32, 64,kernel_size=3,padding=1)
    self.c22 = torch.nn.Linear(64, 64,kernel_size=3,padding=1)

    self.c31 = torch.nn.Linear(64, 128,kernel_size=3,padding=1)
    self.c32 = torch.nn.Linear(128, 128,kernel_size=3,padding=1)
    self.c33 = torch.nn.Linear(128, 128,kernel_size=3,padding=1)

    self.d11 = torch.nn.Linear(192, 64,kernel_size=3,padding=1)


    self.l11 = torch.nn.Linear(3, 32)
    self.l12 = torch.nn.Linear(32, 32)

    self.l2 = torch.nn.Linear(40, 40, bias=False)
    self.l3 = torch.nn.Linear(60, 60, bias=False)
    self.l4 = torch.nn.Linear(80, 1, bias=False)

  def forward(self, x):
    f = torch.nn.functional.leaky_relu(self.l1(x))
    f = torch.cat([x, f], dim=1)
    f = torch.nn.functional.leaky_relu(self.l2(f))
    f = torch.cat([x, f], dim=1)
    f = torch.nn.functional.leaky_relu(self.l3(f))
    f = torch.cat([x, f], dim=1)
    return self.l4(f)[:, 0]
```
